## Handling Data

#### Imported libraraies/packages

```python
import pandas as pd
```

#### Creating a Pandas Series
Series is a one-dimensional object.

```python
# Creating a series by passing in a list
items = pd.Series(['chairs', 'labels', 'tables'])
items
```

```python
# Creating a series by passing in a list with index
items = pd.Series(['chairs', 'labels', 'tables'], index=['OR 001', 'OR 002', 'OR 003'])
items
```

```python
# Creating a series by passing in a list without the index
names = pd.Series(['Joe', 'Tom', 'Jane'])
names
```

#### Creating a DataFrame
DataFrame is like an Excel table. It is a data structure in a table format comprised of rows and columns.

```python
dframe = {'Name': ['Joe', 'Tom', 'Jane'],
          'Items': ['chairs', 'labels', 'tables']
         }
purchases = pd.DataFrame(dframe)
purchases
```
