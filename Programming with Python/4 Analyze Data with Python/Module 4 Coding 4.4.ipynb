### Are the number of reviews increasing each year?

```python
num_reviews = bs.groupby('Year').agg({'Reviews':'sum'})
num_reviews
```

#### Answer: The number of reviews are *_not_* increasing each year.

### Which book received the most number of reviews?

```python
most_num = bs.sort_values(by = 'Reviews', ascending =  False).head(10)
most_num
```

#### Answer: The book that received the most number of reviews *_Where the Crawdad Sing_*.
