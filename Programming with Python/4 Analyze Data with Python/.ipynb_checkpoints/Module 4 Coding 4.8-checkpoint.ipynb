## From Module 4 Coding 4.4.ipynb

```python
import pandas as pd
bs = pd.read_csv('bestsellers.csv')
bs[bs['Name'].isna()]
```

```python
bs = bs.dropna()
bs.isnull().sum()
```

```python
del bs['Type']
bs.head()
```

### Are the number of reviews increasing each year?

```python
num_reviews = bs.groupby('Year', as_index=False).agg({'Reviews':'sum'})
num_reviews
```

### Which book received the most number of reviews?

```python
most_num = bs.sort_values(by = 'Reviews', ascending =  False).head(10)
most_num
```

## Coding Exercises

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
sb.lineplot(x='Year', y='Reviews', data=num_reviews, marker="o")
plt.title('Number of reviews each year')
plt.xlabel('Year')
plt.ylabel('Reviews')

plt.show()
```

```python
sb.barplot(x='Reviews', y='Name', data=most_num, errorbar=('ci', False), orient = 'h', palette = 'Blues')
plt.title('Most number of reviews')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
