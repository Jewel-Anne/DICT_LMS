#### Function

The function **get_mean** will print the average online buys from the past 5 weeks by the **ave** parameter. The value of **mean** is passed into the function and returns a result when the function is called.

```python
def get_mean(ave):
    print("Your average online buys for the past weeks is ", ave)
```

Below, I list the number of items purchased per week and turn it into an **array**.

```python
import array as arr

week_1 = 19
week_2 = 27
week_3 = 17
week_4 = 24
week_5 = 16

week_array = arr.array('i', [week_1, week_2, week_3, week_4, week_5])
print(week_array)
```

Next, to learn the number of weeks, I use the **len() function**.

```python
n = len(week_array)
print(n)
```

To know the average, we must get the sum of the values using the **sum function**.

```python
get_sum = sum(week_array)
print(get_sum)
```

By getting the mean, we can get the average purchase in a span of 5 weeks. 

```python
mean = get_sum/n
print(mean)
```

Now, the function **get_mean** was called by passing the value of **mean** into the function.

```python
get_mean(mean)
```
