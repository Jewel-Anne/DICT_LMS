```python
# Libraries needed
import numpy as np
from scipy import stats
```

```python
sample_data = [103, 151, 176, 188, 146, 184, 175, 112, 115, 163]

mean = np.mean(sample_data)
median = np.median(sample_data)
variance = round(np.var(sample_data, ddof=1), 2)
stdeviation = round(np.std(sample_data, ddof=1), 2)

range = round(np.ptp(sample_data), 2)
mode = stats.mode(sample_data)

print("Mean: ", mean)
print("Mode: ", mode)
print("Median: ", median)
print("Variance: ", variance)
print("Standard Deviation: ", stdeviation)
print("Range: ", range)
```
