```python
import statistics as stats

test_scores = [89, 90, 91, 90, 85, 91, 88, 80, 77, 81, 76, 85, 87, 88, 83, 95, 88, 82, 85, 84]

mean = stats.mean(test_scores)
median = stats.median(test_scores)
mode = stats.mode(test_scores)

print("The Average Score is: ", mean)
print("The Most Frequent Score is: ", mode)
print("The Middle Score is: ", median)
```
